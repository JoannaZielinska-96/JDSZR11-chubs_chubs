```python
fd.groupby('Country')['Popularity'].nlargest(n=1)
```

```python
#Stworzenie grupowania po krajach
groups = fd.groupby(by=['Country'])
```

```python
# Grupowanie według kraju i znajdowanie maksymalnej wartości popularności
max_popularnosc = fd.groupby('Country')['Popularity'].max()
max_popularnosc
```

```python
# Wyznaczenie tytułu piosenki, która jest najbardziej popularna w danym kraju
max_popularnosc = fd.groupby('Country')['Popularity'].max().apply(lambda x: '{:.2f}'.format(x))
max_popularnosc
```

```python
# Sprawdzenie tytułu najpopularniejszej piosenki w danym kraju
result = fd.groupby('Country').apply(lambda x: x.loc[x['Popularity'].idxmax(), 'Title'])
result
```
