Wybór kraju dla którego stworzymy wykres top 5 gatunków muzycznych

```python
# Wybór kraju wprowadzonego przez użytkownika
Wprowadź_kraj = input("Wprowadź nazwę kraju, dla którego chcesz zobaczyć top 5 gatunków muzycznych: ")
```

```python
# Wybór tylko danych dotyczących wybranego kraju i gatunku muzycznego
nowy_fd = fd.loc[fd['Country'] == Wprowadź_kraj, ['Genre', 'Country']]
```

```python
nowy_fd.head(5)
```

```python
# Grupowanie po kraju i gatunku muzycznym, a następnie obliczanie liczby wystąpień

zliczanie = nowy_fd['Genre'].value_counts() #zlicza nam ile jest wystąpień danej kategorii
zliczanie.columns = ['Genre', 'Counts'] #to zmienia nazwe kolumn na genre i counts 
```

```python
zliczanie
```

```python
sprawdzenie = nowy_fd['Genre'].value_counts().nlargest(5)
```

```python
sprawdzenie
```

```python
top5_counts = nowy_fd['Genre'].value_counts().nlargest(5).reset_index().rename(columns={'Genre': 'Counts', 'index': 'Genre'})
```

```python
top5_counts
```

```python
# Tworzenie wykresu słupkowego

plt.figure(figsize=(10,6))
plt.bar(top5_counts['Genre'], top5_counts['Counts'])

plt.title(f"Top 15 gatunków muzycznych według popularności - {Wprowadź_kraj}")

plt.xlabel("Gatunki muzyczne")
plt.ylabel("Liczba wystąpień")

plt.xticks(rotation=90)
plt.show()
```
