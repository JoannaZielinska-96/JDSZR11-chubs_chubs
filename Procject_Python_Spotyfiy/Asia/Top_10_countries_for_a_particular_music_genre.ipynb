```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
pd.set_option('display.max_columns', 151)
```

```python
fd = pd.read_csv('Final database.csv', dtype='unicode')
```

```python
fd.Genre.unique()
```

```python
# Wybór kraju wprowadzonego przez użytkownika
wprowadzony_gatunek = input("Wprowadź nazwę gatunku muzycznego, dla którego chcesz zobaczyć top 10 krajów: ")
```

```python
# Wybór tylko danych dotyczących wybranego kraju i gatunku muzycznego
nowy_fd = fd.loc[fd['Genre'] == wprowadzony_gatunek, ['Genre', 'Country']]
```

```python
# Grupowanie po kraju i gatunku muzycznym, a następnie obliczanie liczby wystąpień

zliczanie = nowy_fd['Country'].value_counts() #zlicza nam ile jest wystąpień danego kraju
zliczanie.columns = ['Country', 'Counts'] #zmienia nazwe kolumn na country i counts 
```

```python
nowy_fd['Country'].value_counts().nlargest(5).reset_index().rename(columns={'Country': 'Counts', 'index': 'Country'})
```

```python
top5_counts = nowy_fd['Country'].value_counts().nlargest(10).reset_index().rename(columns={'Country': 'Counts', 'index': 'Country'})
```

```python
plt.figure(figsize=(10,6))
plt.bar(top5_counts['Country'], top5_counts['Counts'])

plt.title(f"Top 10 krajów według liczby wystąpień gatunku muzycznego - {wprowadzony_gatunek}")

plt.xlabel("Kraje")
plt.ylabel("Liczba wystąpień")

plt.xticks(rotation=90)
plt.show()
```
