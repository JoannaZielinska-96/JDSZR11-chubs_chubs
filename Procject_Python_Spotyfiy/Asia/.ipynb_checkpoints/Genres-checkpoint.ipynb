```python
#Wczytanie bazy danych z wybranymi kolumnami
fd = pd.read_csv('Final database.csv', usecols=['Country', 'Popularity', 'Genre'])
```

```python
#Wy≈õwietlenie 20 przyk≈Çadowych wierszy
fd.sample(20)
```

Sprawdzenie danych

```python
#Sprawdzanie ile jest warto≈õci:

fd.count()
```

```python
#Ile jest unikalnych gatunk√≥w muzycznych w naszej bazie?

unique_genre_count = fd['Genre'].nunique()
unique_genre_count
```

```python
#Ile jest wystƒÖpie≈Ñ r√≥wnych gatunk√≥w muzycznych ≈ÇƒÖcznie w danym kraju?

count_genre2 = fd.groupby('Country')['Genre'].nunique()
count_genre2
```

```python
# W jakim kraju jest najwiƒôksze zr√≥≈ºnicowanie muzyczne? Top 5

unique_genre_counts = fd.groupby('Country')['Genre'].nunique().nlargest(5)
unique_genre_counts
```

```python
plt.title('Top 5 kraj√≥w z najwiƒôkszym zr√≥≈ºnicowaniem muzycznym ')
plt.xlabel('Kraj')
plt.ylabel('Liczba r√≥≈ºnych gatunk√≥w')

unique_genre_counts = fd.groupby('Country')['Genre'].nunique().nlargest(5)
plt.bar(unique_genre_counts.index, unique_genre_counts.values)
```

```python
#Jaki gatunek jest najczƒô≈õciej s≈Çuchany w danym kraju? Licza wystƒÖpie≈Ñ
result = fd.groupby('Country')['Genre'].apply(lambda x: x.value_counts().nlargest(1))
result
```
